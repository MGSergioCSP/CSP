### Exercicio 2: Operadores de Asignación

Simula unha conta bancaria:
- Saldo inicial: 1000
- Usa += para depositar 500
- Usa -= para retirar 300
- Usa *= para aplicar interés do 10%
- Usa /= para dividir entre 2 (metade para aforros)
- Amosa saldo despois de cada operación
- Utiliza diferentes formatos de cadeas.

```python
# Pon a túa solución aquí
saldo = float(input("Introduce o saldo inicial da conta: "))
print(f"Saldo inicial: {saldo}€")

deposito = float(input("Canto desexas depositar? "))
saldo += deposito
print("Despois de depositar {}, o saldo é de: {}€".format(deposito, saldo))

retirada = float(input("Canto desexas retirar? "))
saldo -= retirada
print("Despois de retirar %d, o saldo queda en: %.2f€" % (retirada, saldo))

interes_pct = 10
print("Aplicarase un 10% de interese ")

saldo *= (1 + interes_pct / 100)
print(f"Tras aplicar un {interes_pct}% de interés, o saldo aumenta a: {saldo:.2f}€")

print("Dividindo o saldo á metade para a conta de aforros...")
saldo /= 2
print("O saldo final na conta corrente é: {0:.2f}€".format(saldo))


```

### Exercicio 8: Walrus

- Lonxitude >= 8 caracteres.
- Contén maiúscula e número.
- Usa walrus para non repetir len(contrasinal).

```python
# Pon a túa solución aquí
while (
    (contrasinal := input("\nIntroduce unha nova clave: ")) and
    ((n := len(contrasinal)) < 8 or
     contrasinal.islower() or
     contrasinal.isalpha())
):
    print(f"Clave insegura (Longitud: {n}).")
    print("Requisitos: Mínimo 8 caracteres, una mayuscula y un número.")

print(f"\Contraseña aceptada! Longitud final: {n}")
```

### Exercicio 9: Combinación de Operadores

Calculadora de IMC (xa a fixemos pero desta volta é máis completa):
- peso / altura²
- Clasifica segundo táboa OMS
- Baixo peso: < 18.5
- Normal: 18.5 - 24.9
- Sobrepeso: 25 - 29.9
- Obesidade: >= 30
- Usa operadores de comparación e lóxicos
- Opcional: walrus para evitar repetir cálculo

```python
# Pon a túa solución aquí
# Entrada de datos
peso = float(input("Introduce tu peso en kg: "))
altura = float(input("Introduce tu altura en metros: "))

if (imc := peso / (altura ** 2)) < 18.5:
    clasificacion = "Bajo peso"
elif 18.5 <= imc <= 24.9:
    clasificacion = "Normal"
elif 25 <= imc <= 29.9:
    clasificacion = "Sobrepeso"
else:
    clasificacion = "Obesidad"

print(f"\nTu IMC es: {imc:.2f}")
print(f"Clasificación OMS: {clasificacion}")
```

### Exercicio 10: Analizador de Texto

Analiza un texto introducido polo usuario:
1. Lonxitude.
2. Se contén certas palabras.
3. Conta vogais.
4. Verifica se é palíndromo.
5. Verifica se texto é `None`.
6. Concatena.

```python
# Pon a túa solución aquí
# Entrada de datos
texto = input("Introduce un texto para analizar: ")

if texto is None:
    print("El texto es nulo. No se puede analizar.")
else:
    longitud = len(texto)

    vocales = "aeiouáéíóú"
    num_vocales = sum(1 for letra in texto.lower() if letra in vocales)

    palabra_buscar = input("Palabra a buscar: ")
    contiene_palabra = palabra_buscar.lower() in texto.lower()

    texto_limpio = texto.replace(" ", "").lower()
    is_palindromo = texto_limpio == texto_limpio[::-1] if texto_limpio else False

    texto_final = texto + " [ANALIZADO]"

    # --- RESULTADOS ---
    print("\n" + "="*20)
    print(f"ANÁLISIS DE: '{texto}'")
    print("="*20)
    print(f"* Longitud: {longitud} caracteres")
    print(f"* Número de vocales: {num_vocales}")
    print(f"* Coniene la palabra '{palabra_buscar}'?: {'Si' if contiene_palabra else 'No'}")
    print(f"* Es palíndromo?: {'Si' if is_palindromo else 'No'}")
    print(f"* Texto concatenado: {texto_final}")
```

### Xogo de Adiviñanza

Xogo onde:
- Número secreto aleatorio (1-100).
- Aportar Pistas ("é maior", "é menor").
- Contador de intentos.
- Utiliza Walrus.
- Preguntar se se quere xogar outra vez.

```python
# Pon a túa solución aquí
import random

print("Bienvenido a 'Adivina el número'!")

numero_secreto = random.randint(1, 100)

intentos = 0
adivinado = False

while not adivinado:
    suposicion = input("Ingresa tu suposición (entre 1 y 100): ")

    if not suposicion.isdigit():
        print("Por favor, ingresa un número válido.")
        continue

    suposicion = int(suposicion)

    if suposicion < 1 or suposicion > 100:
        print("El número debe estar en el rango de 1 a 100.")
        continue
    intentos += 1

    if suposicion < numero_secreto:
        print("Demasiado bajo. ¡Intenta nuevamente!")
    elif suposicion > numero_secreto:
        print("Demasiado alto. ¡Intenta nuevamente!")
    else:
        print(f"¡Felicidades! ¡Adivinaste el número en {intentos} intentos!")
        adivinado = True

    print(f"Intentos realizados: {intentos}")


print(f"¡Felicidades! ¡Adivinaste el número en {intentos} intentos!")
adivinado = True

jugar_nuevamente = input("¿Deseas jugar nuevamente? (sí/no): ")

if jugar_nuevamente.lower() == 'sí':
    numero_secreto = random.randint(1, 100)
    intentos = 0
    adivinado = False
    print("¡Genial! ¡Vamos a jugar nuevamente!")
else:
    print("Gracias por jugar. ¡Hasta la próxima!")
```

### Validador de Email

Valida email con:
- Que conteña unha `@`.
- Que conteña un `.` despois da `@`.
- Lonxitude >= 5.
- Non conteña espazos en branco.
- Dominio pertenza a unha lista de dominios válida.

```python
# Pon a túa solución aquí
# Lista de dominios permitidos
dominios_validos = ["gmail.com", "outlook.com", "yahoo.es", "udc.gal", "hotmail.com"]

print("--- Validador de Correo Electrónico ---")

while True:
    email = input("\nIntroduce tu email para validar: ").strip()

    if len(email) < 5:
        print("Error: El email es demasiado corto (mínimo 5 caracteres).")
        continue

    if " " in email:
        print("Error: El email no puede contener espacios en blanco.")
        continue

    if "@" not in email:
        print("Error: El email debe contener una '@'.")
        continue

    partes = email.split("@")

    if len(partes) != 2:
        print("Error: El email debe tener exactamente una '@'.")
        continue

    usuario, dominio = partes

    if "." not in dominio:
        print("Error: Debe haber un punto (.) después de la '@'.")
        continue

    if dominio not in dominios_validos:
        print(f"Error: El dominio '{dominio}' no está permitido.")
        print(f"Dominios válidos: {', '.join(dominios_validos)}")
        continue

    print(f"¡Éxito! El email '{email}' es válido.")
    break
```
